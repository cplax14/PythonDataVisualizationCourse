```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#seaborn comes with some sample datasets, such as the tips dataset
tips = sns.load_dataset('tips')
#this loads the data in as a data frame
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
#does a scatter plot of the first column to the second, and adds the linear regression line of their relationship
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
#the trend line is only a first order trend line
```

```python
sns.lmplot('total_bill','tip',tips,order=3,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
#can change the order
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
#by removing the fig_reg, you just get the scatter plot
```

```python
tips.head()
```

```python
tips['tip_pct']=100*(tips['tip']/tips['total_bill'])
#making a new column on the data frame for the tip percentage
tips.head()
```

```python
sns.lmplot('size','tip_pct',tips)
```

```python
#check out what a jitter is
sns.lmplot('size','tip_pct',tips,x_jitter=.1)
#this spreads the data out a little bit to make the data easier to read
```

```python
sns.lmplot('size','tip_pct',tips,x_estimator=np.mean)
#shows the tendency for each bin 
```

```python
#hue facet to automatically define subsets along the column
sns.lmplot('total_bill','tip_pct',tips,hue='sex',markers=['x','o'])
#lets you put two linear regression plots on top of each other
```

```python
#does the day make a difference in how much the tip is
sns.lmplot('total_bill','tip_pct',tips,hue='day')
#by leaving the default for hue without markers, seaborn will do the default, which uses colors
```

```python
#seaborn support local regression model fitting, an advanced statistical model fitting or LOES
sns.lmplot('total_bill','tip_pct',tips,lowess=True,line_kws={'color':'black'})
```

```python
#lmplot uses regplot
sns.regplot('total_bill','tip_pct',tips)
```

```python
sns.lmplot('total_bill','tip_pct',tips)
```

```python
#can set subplots
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)
#1 row and 2 columns
sns.regplot('total_bill','tip_pct',tips,ax=axis1)
sns.violinplot(tips['tip_pct'],tips['size'],color='Reds_r',ax=axis2)
```
