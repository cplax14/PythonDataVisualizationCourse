```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.distplot(dataset,bins=25)
#allows us to combine plots
#bins are set to 10 by default in seaborn
#shows histogram and kde plot by default
```

```python
sns.distplot(dataset,bins=25,rug=True,hist=False)
#only shows kde plot and rug plot
```

```python
#can pass a dictionary to modify both plots at once
sns.distplot(dataset,bins=25,
            kde_kws={'color':'indianred','label':'KDE PLOT'},
            hist_kws={'color':'blue','label':'HIST PLOT'})
```

```python
from pandas import Series

ser1 = Series(dataset,name='My_data')
ser1
```

```python
sns.distplot(ser1,bins=25)
#can plot a pandas series
```
