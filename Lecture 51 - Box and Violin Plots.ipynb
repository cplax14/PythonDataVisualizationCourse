```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
#whis means whiskers, and we're setting it to the numpy value of infinity. This allow it to include outliers
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data1,vert=False)
#allows you to do a horizontal box plot
```

```python
#Normal distribution from 0 to 5
data1 = stats.norm(0,5).rvs(100)

#two gamma distributions concatenated
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)-1])

#box plot both data1 and data2
sns.boxplot([data1,data1],whis=np.inf)
```

```python
#they both look very similar on a box plot, which is where a violin plot can come in handy by bringing in the KDE
sns.violinplot([data1,data2])
#you can see here, with one positive gamma distribution, and one negative distribution
```

```python
sns.violinplot(data2,bw=0.01)
#0.01 is a very extreme bandwidth
```

```python
sns.violinplot(data1,inner='stick')
#gives you a rug plot in the violin plot
```
